### üì• Load Raw Dataset

### üéØ Target Variable Distribution

Before diving into feature relationships, let's understand the target variable ‚Äî the loan repayment status.

This gives us a quick view of class balance and helps us decide if we need to address imbalance during modeling.

```python
# Plotting class distribution of the target variable (0 = repaid, 1 = default)
df['TARGET'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title("Distribution of Loan Repayment vs Default")
plt.xlabel("Target Variable (0 = Repaid, 1 = Defaulted)")
plt.ylabel("Number of Applicants")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### ‚ö†Ô∏è Class Imbalance in the Target Variable

The dataset is significantly imbalanced ‚Äî only **~8% of applicants** defaulted (`TARGET = 1`), while **~92%** successfully repaid their loans (`TARGET = 0`).

This mirrors real-world lending patterns, where defaults are rare but financially critical to detect.

##### üß† Why Accuracy Alone Is Misleading

A model could achieve over **90% accuracy** simply by predicting `0` (no default) for every applicant. That might look impressive ‚Äî but it would **miss all actual defaults**, making the model practically useless for risk assessment.

##### ‚úÖ What I‚Äôll Focus On Instead

To evaluate model performance meaningfully, I‚Äôll prioritize metrics that capture the ability to detect rare but important defaults:

- **Recall (Sensitivity):** How well the model catches actual defaulters  
- **AUC-ROC:** Overall classification performance across thresholds  
- **Precision-Recall Curve:** More informative under class imbalance  

These metrics ensure the model can **spot high-risk applicants** even when they‚Äôre few and far between.

### üßº Missing Values & Column Overview

Before diving into modeling, it‚Äôs important to understand the structure of the data ‚Äî especially missingness.

In this step, I generate a quick summary for all columns, including:
- üßæ Data type (numerical, categorical, etc.)

```python
# Create a summary of all columns: data type, % missing values, and number of unique values
summary_df = pd.DataFrame({
    'Data Type': df.dtypes,
    'Missing %': df.isnull().mean().round(4) * 100,
    'Unique Values': df.nunique()
}).sort_values(by='Missing %', ascending=False)

# Preview the top 50 columns with the highest missing percentages
summary_df.head(50)
```

### üìå Dropping Columns with High Missingness or Cardinality

From the column summary above, I noticed several features with:
- **Extremely high missing rates** (many >50%, some over 80%)
- **High cardinality** or sparse, hard-to-interpret values
- **Redundancy** with more usable or interpretable features

To reduce noise and simplify modeling, I applied the following criteria:

##### ‚úÖ Columns were **kept** if they were:
- Relevant to the prediction target (`TARGET`)
- Interpretable by domain experts or stakeholders
- Easy to encode or impute
- Known to have predictive power (e.g., `EXT_SOURCE_1`)
- Demographic or behavioral in nature

##### ‚ùå Columns were **dropped** if:
- They had **>50% missing values** and weak business relevance
- They were difficult to interpret or impute (e.g., rare apartment metadata)
- They had **high cardinality** or few distinct patterns
- They duplicated information available in other features

üß† Why 50%?

This threshold is commonly used in applied ML to reduce noise while preserving meaningful signals. Features with more than half of their values missing often lack stability and business utility ‚Äî especially when there's no domain justification for their absence. This helped reduce the dataset‚Äôs dimensionality and focus attention on features most likely to drive meaningful credit risk predictions.

```python
# List of columns to drop based on prior summary and domain logic
columns_to_drop = [
    'COMMONAREA_MEDI', 'COMMONAREA_AVG', 'COMMONAREA_MODE',
    'NONLIVINGAPARTMENTS_MODE', 'NONLIVINGAPARTMENTS_AVG', 'NONLIVINGAPARTMENTS_MEDI',
    'FONDKAPREMONT_MODE', 'LIVINGAPARTMENTS_MODE', 'LIVINGAPARTMENTS_AVG', 'LIVINGAPARTMENTS_MEDI',
    'FLOORSMIN_AVG', 'FLOORSMIN_MODE', 'FLOORSMIN_MEDI',
    'YEARS_BUILD_MEDI', 'YEARS_BUILD_MODE', 'YEARS_BUILD_AVG',
    'OWN_CAR_AGE', 'LANDAREA_MEDI', 'LANDAREA_MODE', 'LANDAREA_AVG',
    'BASEMENTAREA_MEDI', 'BASEMENTAREA_AVG', 'BASEMENTAREA_MODE',
    'NONLIVINGAREA_MODE', 'NONLIVINGAREA_AVG', 'NONLIVINGAREA_MEDI',
    'ELEVATORS_MEDI', 'ELEVATORS_AVG', 'ELEVATORS_MODE',
    'APARTMENTS_MEDI', 'APARTMENTS_AVG', 'APARTMENTS_MODE',
    'ENTRANCES_MEDI', 'ENTRANCES_AVG', 'ENTRANCES_MODE',
    'LIVINGAREA_AVG', 'LIVINGAREA_MODE', 'LIVINGAREA_MEDI',
    'TOTALAREA_MODE'
]

# Drop the selected columns from the dataframe
df = df.drop(columns=columns_to_drop)

# Display how many columns remain
print(f"Remaining columns: {df.shape[1]}")
```

### üßπ Missing Value Treatment Strategy

To prepare the dataset for modeling, I analyzed the remaining missing values and applied the following structured approach:

##### üîª Dropped Columns
Removed columns with **>50% missing values** and **low interpretability or business value**, such as:
- `WALLSMATERIAL_MODE`, `FLOORSMAX_*`, `YEARS_BEGINEXPLUATATION_MODE/MEDI`

These features were unlikely to contribute meaningfully to prediction and often contained sparse or noisy information.

##### üîß Imputed Numerical Columns (with Median)
Used **median imputation** for key numerical features where missingness was common but values were meaningful:
- External risk scores: `EXT_SOURCE_1`, `EXT_SOURCE_2`, `EXT_SOURCE_3`
- Building age: `YEARS_BEGINEXPLUATATION_AVG`
- Financial indicators: `AMT_GOODS_PRICE`, `AMT_ANNUITY`
- Demographics: `CNT_FAM_MEMBERS`, `DAYS_LAST_PHONE_CHANGE`

This choice helps maintain robust central tendencies without being overly influenced by outliers.

##### üßä Filled Zero Where Missing Implies Absence
For some features, missing values likely indicate the absence of behavior or relationships. In these cases, I filled with `0`:
- Credit behavior: `AMT_REQ_CREDIT_BUREAU_*`
- Social metrics: `DEF_30_CNT_SOCIAL_CIRCLE`, `OBS_30_CNT_SOCIAL_CIRCLE`, etc.

This preserves the signal while reflecting real-world meaning behind the absence.

##### üè∑Ô∏è Imputed Categorical Columns with ‚ÄúMissing‚Äù Label
Low-cardinality categorical features were imputed using a `"Missing"` category:
- Examples include: `HOUSETYPE_MODE`, `EMERGENCYSTATE_MODE`, `OCCUPATION_TYPE`, `NAME_TYPE_SUITE`

This keeps all rows usable while preserving interpretability for downstream modeling and SHAP analysis.

#### üîç Why Not Drop Everything with >50% Missing?

üëâ Some features (like `EXT_SOURCE_1`) had **high missingness but strong predictive power**. I chose to **retain and impute** these carefully using median or business-informed defaults.

üëâ This reflects a **balanced trade-off** between data quality and model utility ‚Äî not just a blanket threshold rule.

This helped prioritize features for encoding and model input.

```python
# Drop selected columns with high missingness and low predictive value
cols_to_drop = [
    'WALLSMATERIAL_MODE',
    'FLOORSMAX_AVG', 'FLOORSMAX_MEDI', 'FLOORSMAX_MODE',
    'YEARS_BEGINEXPLUATATION_MEDI', 'YEARS_BEGINEXPLUATATION_MODE'
]
df.drop(columns=cols_to_drop, inplace=True)

# Impute missing values in numerical columns using the median
median_impute_cols = [
    'EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3',
    'YEARS_BEGINEXPLUATATION_AVG', 'AMT_GOODS_PRICE', 'AMT_ANNUITY',
    'CNT_FAM_MEMBERS', 'DAYS_LAST_PHONE_CHANGE'
]
for col in median_impute_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill missing values with 0 where null indicates absence of behavior or event
zero_fill_cols = [
    'AMT_REQ_CREDIT_BUREAU_HOUR', 'AMT_REQ_CREDIT_BUREAU_DAY',
    'AMT_REQ_CREDIT_BUREAU_WEEK', 'AMT_REQ_CREDIT_BUREAU_MON',
    'AMT_REQ_CREDIT_BUREAU_QRT', 'AMT_REQ_CREDIT_BUREAU_YEAR',
    'DEF_30_CNT_SOCIAL_CIRCLE', 'DEF_60_CNT_SOCIAL_CIRCLE',
    'OBS_30_CNT_SOCIAL_CIRCLE', 'OBS_60_CNT_SOCIAL_CIRCLE'
]
df[zero_fill_cols] = df[zero_fill_cols].fillna(0)

# Fill missing values in categorical variables with the label "Missing"
cat_fill_cols = ['HOUSETYPE_MODE', 'EMERGENCYSTATE_MODE', 'OCCUPATION_TYPE', 'NAME_TYPE_SUITE']
df[cat_fill_cols] = df[cat_fill_cols].fillna('Missing')

# Print shape to confirm row and column count after imputations
df.shape
```

### üßπ Removing Low-Value and Redundant Columns

To further streamline the dataset and reduce noise, I removed several groups of features with little or no predictive value:

- **Low-variance columns**: Features with only one unique value (i.e., same for all applicants) provide no information for modeling.
- **Document flag columns**: The `FLAG_DOCUMENT_*` features indicate submission of paperwork types, but all had minimal variance and unclear predictive value.
- **Unique identifier**: The `SK_ID_CURR` column is a unique ID with no modeling relevance.

These removals simplify the feature set and reduce dimensionality before encoding and modeling.

```python
# Drop SK_ID_CURR (just an ID), all FLAG_DOCUMENT_* columns, and any columns with only 1 unique value

# Identify columns with only one unique value (constant)
low_variance_cols = [col for col in df.columns if df[col].nunique() == 1]

# Identify document flag columns
doc_flag_cols = [col for col in df.columns if col.startswith('FLAG_DOCUMENT_')]

# Identify ID column (not useful for modeling)
id_col = ['SK_ID_CURR']

# Combine all columns to drop
drop_cols = low_variance_cols + doc_flag_cols + id_col

# Drop selected columns from dataframe
df.drop(columns=drop_cols, inplace=True)

# Display remaining number of columns
print(f"Dropped {len(drop_cols)} columns. Remaining shape: {df.shape}")
```

At this stage, the dataset has been fully cleaned:

- All missing values have been handled through a combination of dropping and imputation.
- Low-variance, redundant, or ID columns have been removed.

## üî° Encoding Categorical Variables for Modeling

Now that the dataset is cleaned and imputed, it‚Äôs time to transform categorical features into numerical formats that machine learning models can understand.

I used two encoding techniques based on the **cardinality** of each feature:

#### üß© One-Hot Encoding for Low-Cardinality Features

These are features with a small number of distinct categories ‚Äî ideal for one-hot encoding, which expands them into binary indicator columns.

Applied to:
- `NAME_CONTRACT_TYPE`, `FLAG_OWN_CAR`, `FLAG_OWN_REALTY`, `CODE_GENDER`
- `EMERGENCYSTATE_MODE`, `HOUSETYPE_MODE`, `NAME_EDUCATION_TYPE`
- `NAME_FAMILY_STATUS`, `NAME_HOUSING_TYPE`, `WEEKDAY_APPR_PROCESS_START`
- `NAME_TYPE_SUITE`, `NAME_INCOME_TYPE`

**Why One-Hot?**
- Prevents introducing spurious relationships with the target
- Keeps interpretability and works well for tree-based and linear models

#### üß† Target Encoding for High-Cardinality Features (with Cross-Validation)

Used for:
- `OCCUPATION_TYPE`
- `ORGANIZATION_TYPE`

Target encoding replaces each category with the **mean target value**, but to prevent data leakage, I applied it **fold-wise using cross-validation**:
- For each fold, only training data was used to compute target means
- This prevents overfitting and leakage from target values in the validation fold

**Why Target Encoding?**
- One-hot encoding would create too many sparse columns
- Target encoding preserves signal while staying compact

#### ‚úÖ Result:
- Original categorical columns were encoded and dropped
- Final encoded dataset (`df_encoded`) is now fully numeric
- Shape after encoding: **88 features √ó 307,511 rows**

This dataset is now ready for model training, evaluation, and SHAP analysis.

```python
# Identify all categorical columns based on data type
categorical_cols = df.select_dtypes(include='object').columns.tolist()

# Display the number of unique categories for each categorical column (sorted by count)
df[categorical_cols].nunique().sort_values()
```

```python
# Create a copy of the cleaned dataset for encoding (preserve original)
df_encoded = df.copy()

# One-hot encode low-cardinality categorical features
low_cardinality_cols = [
    'NAME_CONTRACT_TYPE', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 'CODE_GENDER',
    'EMERGENCYSTATE_MODE', 'HOUSETYPE_MODE', 'NAME_EDUCATION_TYPE',
    'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE', 'WEEKDAY_APPR_PROCESS_START',
    'NAME_TYPE_SUITE', 'NAME_INCOME_TYPE'
]

# Apply one-hot encoding and drop the first category to avoid multicollinearity
df_encoded = pd.get_dummies(df_encoded, columns=low_cardinality_cols, drop_first=True)

# Target encode high-cardinality categorical features using cross-validation
high_cardinality_cols = ['OCCUPATION_TYPE', 'ORGANIZATION_TYPE']
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Encode each column fold-by-fold to avoid target leakage
for col in high_cardinality_cols:
    df_encoded[col + '_TE'] = np.nan  # initialize encoded column
    for train_idx, valid_idx in kf.split(df_encoded):
        te = TargetEncoder(cols=[col])
        encoded = (
            te.fit(df_encoded.iloc[train_idx], df_encoded.iloc[train_idx]['TARGET'])
              .transform(df_encoded.iloc[valid_idx])[col]
              .astype(float)  # cast to float to avoid dtype issues
        )
        df_encoded.loc[df_encoded.index[valid_idx], col + '_TE'] = encoded

# Drop the original categorical columns after encoding
df_encoded.drop(['ORGANIZATION_TYPE', 'OCCUPATION_TYPE'], axis=1, inplace=True)

# Print final shape and preview the encoded DataFrame
print("New shape after encoding:", df_encoded.shape)
print("Preview of encoded data:")
display(df_encoded.head())
```

### üíæ Save Cleaned + Encoded Dataset

To keep the workflow modular, I saved the fully cleaned and encoded dataset to disk.  
This allows me to load it directly in the modeling notebook without repeating earlier steps.

Before saving, I also re-cleaned the column names to ensure they are fully compatible with modeling libraries like LightGBM, which can fail if special characters (e.g. `/`, `,`, `(`) are present in feature names.

```python
# Clean column names one last time to ensure modeling compatibility
df_encoded.columns = df_encoded.columns.str.replace(r"[^A-Za-z0-9_]+", "_", regex=True)

# Save the final encoded dataset to disk for use in modeling
df_encoded.to_parquet("../data/processed/application_train_encoded.parquet", index=False)
```
